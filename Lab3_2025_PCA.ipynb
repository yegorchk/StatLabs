```python
# Центрирование данных

def center_data(X, Y):
  """Центрирование данных.

  Параметры
  ----------
  X : {array-like} формы (n_samples, n_features)
      Центрируемый набор данных, где `n_samples` — число объектов
      и `n_features` — число признаков.
  Y : {array-like} формы (n_samples, n_features)
      Набор данных, относительно которого производится центрирование,
      где `n_samples` — число объектов
      и `n_features` — число признаков.

  Что возвращается
  -------
  X_new : массив формы (n_samples, n_features) — преобразованные данные.
  """
  return X - X.mean()
  pass

data_train_centered =  center_data(data_train, data_train)

assert np.all(np.isclose(np.sum(center_data(data_train, data_train),axis=0),0))
```
